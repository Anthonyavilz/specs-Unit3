### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
number_df = pd.read_csv('salary_numeric.csv')
```

```python
number_df.head()
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = The higher education level you have, the higher the base salary and overall salary you make is.

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(number_df['salary'], number_df['education'])
```

- With the correlation being a positive number, it shows that theres an overall positive relationship with the data

#### Are these results statistically significant? How do you know?

- There is a statistical significance with the result as the p-value shows a 0.

### What is the correlation between salary and age?

#### Your Hypothesis = The older you get the higher salary you get

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(number_df['salary'], number_df['age'])
```

- The correlation shows a positive number so there is an overall positive relationship with the data

#### Are these results statistically significant? How do you know? 

- With the p-value being 0, there is a statistical significance with the data

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = The higher amount of industry experience, the higher amount of salary you earn.

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(number_df['salary'], number_df['industry_experience'])
```

- The correlation shows a positive number so the the relationship in the data will be an upward trend.

#### Are these results statistically significant? How do you know?

- Like the previous findings, the p-value is 0 meaning there is statistical significance.
